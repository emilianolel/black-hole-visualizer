# 🌌 Schwarzschild Physics Regression Visualizer

This notebook provides a cinematic and mathematical verification of the Schwarzschild Geodesic Integrator. We visualize the stability of the **Photon Sphere** (at $3M$) and the gravitational capture at the **Event Horizon** (at $2M$).

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

# Ensure the src directory is in the path for importing the integrator
sys.path.append(os.path.abspath('..'))
from src.engine.integrator import trace_photon, RS, M

print(f"Schwarzschild Constants: M={M}, RS={RS}")
```

## 1. Case: Photon Sphere Stability ($r = 3M$)

In the Schwarzschild metric, $r = 3M$ is the uniquely unstable circular orbit for photons. A perfectly balanced photon should maintain its radius, while numerical drift will eventually cause it to spiral in or escape.

```python
# Define circular null geodesic at r=3M
r0 = 3.0 * M
theta0 = np.pi / 2.0
phi0 = 0.0
pt0 = -1.0
pr0 = 0.0
ptheta0 = 0.0
pphi0 = 3.0 * np.sqrt(3.0) * M

initial_state = np.array([0.0, r0, theta0, phi0, pt0, pr0, ptheta0, pphi0])

# Trace for several orbits
path = trace_photon(initial_state, step_size=0.1, max_steps=1000)

# Extract Cartesian coordinates for plotting
r_vals = path[:, 1]
theta_vals = path[:, 2]
phi_vals = path[:, 3]

x = r_vals * np.sin(theta_vals) * np.cos(phi_vals)
y = r_vals * np.sin(theta_vals) * np.sin(phi_vals)

plt.figure(figsize=(8, 8), facecolor='#050505')
ax = plt.gca()
ax.set_facecolor('#050505')

# Plot Event Horizon (RS=2.0)
horizon = plt.Circle((0, 0), RS, color='#ff0000', alpha=0.3, label="Event Horizon (2M)")
ax.add_patch(horizon)

# Plot Photon Sphere Boundary (r=3.0)
sphere_boundary = plt.Circle((0, 0), 3.0, color='#00ffff', fill=False, linestyle='--', alpha=0.5, label="Photon Sphere (3M)")
ax.add_patch(sphere_boundary)

# Plot Geodesic Trace
plt.plot(x, y, color='#00ffaa', linewidth=2, label="Photon Path")
plt.scatter(x[0], y[0], color='white', zorder=5, label="Start")

plt.title("Photon Sphere Stability Test", color='white', fontsize=14)
plt.xlabel("x (units of M)", color='#888')
plt.ylabel("y (units of M)", color='#888')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.legend(facecolor='#111', edgecolor='#333', labelcolor='white')
plt.grid(alpha=0.1)
plt.show()
```

## 2. Case: Gravitational Capture ($r \le 2M$)

Here we launch a photon with negative radial momentum toward the black hole to verify that the integrator terminates correctly at the event horizon boundary.

```python
r0 = 10.0 * M
initial_state = np.array([0.0, r0, np.pi/2, 0.0, -1.0, -0.6, 0.0, 0.2]) # High radial momentum + slight spiral

path = trace_photon(initial_state, step_size=0.1, max_steps=1000)

r_vals = path[:, 1]
theta_vals = path[:, 2]
phi_vals = path[:, 3]

x = r_vals * np.sin(theta_vals) * np.cos(phi_vals)
y = r_vals * np.sin(theta_vals) * np.sin(phi_vals)

plt.figure(figsize=(8, 8), facecolor='#050505')
ax = plt.gca()
ax.set_facecolor('#050505')

plt.Circle((0, 0), RS, color='#ff0000', alpha=0.3)
ax.add_patch(plt.Circle((0, 0), RS, color='#ff0000', alpha=0.3, label="Event Horizon"))

plt.plot(x, y, color='#ffaa00', linewidth=2, label="Captured Photon")
plt.scatter(x[0], y[0], color='white', zorder=5, label="Start (r=10M)")

plt.title("Event Horizon Capture Test", color='white', fontsize=14)
plt.xlabel("x (units of M)", color='#888')
plt.ylabel("y (units of M)", color='#888')
plt.xlim(-12, 12)
plt.ylim(-12, 12)
plt.legend(facecolor='#111', edgecolor='#333', labelcolor='white')
plt.grid(alpha=0.1)
plt.show()
```
