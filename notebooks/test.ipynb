```python
# Bucket identificado de tu infraestructura
bucket_name = "black-hole-visualizer-project-bh-vis-dataproc-config" 

# 1. Crear datos de prueba
data = [("Evento de Horizonte", 10.5), ("Disco de Acreción", 150.2), ("Jet Relativista", 99.9)]
columns = ["Componente", "Energía_Relativa"]

df = spark.createDataFrame(data, columns)

print(f"--- Escribiendo datos en gs://{bucket_name}/test_connection/ ---")

# 2. Escribir en GCS (formato Parquet, que es más eficiente para Spark)
df.write.mode("overwrite").parquet(f"gs://{bucket_name}/test_connection/")

print("--- Lectura exitosa desde GCS: ---")

# 3. Leer de vuelta para verificar integridad
df_verify = spark.read.parquet(f"gs://{bucket_name}/test_connection/")
df_verify.show()

print("¡Prueba completada con éxito! El cluster tiene lectura/escritura total en GCS.")

```
